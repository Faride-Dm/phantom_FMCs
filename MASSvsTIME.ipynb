## This is a script to plot mass in solar mass unit against time in Kyr using FUNCTIONS for jobs that have more than two sinks

```python
import numpy as np
import math
from astropy import constants as const
from astropy import units as u
```

```python
import matplotlib.pyplot as plt
%matplotlib nbagg
import matplotlib.ticker as mtick
# plt.style.use('seaborn-white')
```

```python
timeUnitSi = (u.m**3/const.G.unit/u.kg)**(0.5); timeUnitSi
```

```python
timeUnit = (const.pc**3/(const.G)/const.M_sun)**(0.5); timeUnit
```

```python
def openfile (sinkFilePath):
    print("Opening " + sinkFilePath)
    time, mass = np.loadtxt(sinkFilePath,unpack=True, usecols=(0,4))
    time=time*timeUnit # time in second
    time = time/u.yr.to('s')/1000 # time in kyr
#     time=time/(31.5576*10**9)
    return time, mass
```
